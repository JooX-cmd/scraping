```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python

date = input("Enter the date (MM/DD/YYYY): ") 
page  = requests.get(f"https://www.yallakora.com/match-center?date={date}#days")


page_content = BeautifulSoup(page.content, "html.parser")

page_content

matches_details = []

championships = page_content.find_all("div", class_="matchCard")

```

## Get all matches



```python
with open("matches_details.csv", "w", newline="", encoding="utf-8") as csvfile:
    fieldnames = ["Championship", "Week", "Team 1", "Team 1 Score", "Team 2", "Team 2 Score", "Final Score", "Kick-off Time", "Status", "Channel"]
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    
    writer.writeheader()
    
    for championship in championships:
        championship_title = championship.find("h2").text.strip()
        
        matches = championship.find_all("div", class_="item")
        
        for match in matches:
            try:
                team_a = match.find("div", class_="teams teamA")
                team1_name = team_a.find("p").text.strip() 
                
                team_b = match.find("div", class_="teams teamB")
                team2_name = team_b.find("p").text.strip()
                
                status_div = match.find("div", class_="matchStatus")
                status = status_div.find("span").text.strip() 
                
                scores = match.find_all("span", class_="score")
                team1_score = scores[0].text.strip()
                team2_score = scores[1].text.strip() 
                final_score = f"{team1_score} - {team2_score}"
                
                time_span = match.find("span", class_="time")
                time = time_span.text.strip() 
                
                channel_div = match.find("div", class_="channel")
                channel = channel_div.text.strip()
                
                date_div = match.find("div", class_="date")
                match_week = date_div.text.strip() 
                
                writer.writerow({
                    "Championship": championship_title,
                    "Week": match_week,
                    "Team 1": team1_name,
                    "Team 1 Score": team1_score,
                    "Team 2": team2_name,
                    "Team 2 Score": team2_score,
                    "Final Score": final_score,
                    "Kick-off Time": time,
                    "Status": status,
                    "Channel": channel
                })
                
                print(f" {team1_name} vs {team2_name}")
                
            except Exception as e:
                print(f"Error processing match: {e}")

```
