```python
jobs_data = open("jobs_data.txt", "w")
```

```python
import csv
writer = csv.writer(jobs_data)
writer.writerow(["Job Title", "Company Name", "Location", "Job Type"])
for i in range(0,21):
    url = f"https://wuzzuf.net/search/jobs?q=ai+engenner&a=navbl&start={i}"
    page_content = requests.get(url).text
    website  =  BeautifulSoup(page_content, "html")
    jops = website.find_all("div", class_ = "css-pkv5jc")
    for job in jops:
        jop_title = job.find("a", class_="css-o171kl").text
        company_name = job.find("a", class_="css-ipsyv7").text
        location = job.find("span", class_="css-16x61xq").text
        jop_type = job.find("span", class_="css-uc9rga eoyjyou0").text
        
        writer.writerow([jop_title, company_name, location, jop_type])
```
